## Bifurcation Diagram

```python
from math import sin, pi, log, e

import seaborn as sns; sns.set()


N = 100
x = 0
x_data = []
alpha_data = []
alpha_bin = .001
alpha_tot= 1000
x_tot = 10
x_bin = 1 / x_tot
for alpha in range(400, alpha_tot):
    for x in range(1, x_tot):
        alpha_data.append(float(alpha_bin) * (alpha))
        xcurr = x * x_bin        
        for i in range(N):
            xcurr = float(alpha_bin) * (alpha)*sin(pi*xcurr)
        x_data.append(xcurr)

plt.scatter(alpha_data, x_data, s=pi / 2)
plt.ylabel('x')
plt.xlabel('alpha')
```

## Variability in x at alpha=.9

```python
N = 1000
xcurr = .4
alpha2 = .9
epsilon2 = .001
xcurr2 = xcurr + epsilon2
xs = []
for i in range(N):
    if i > 3:
        xs.append((xcurr2-xcurr)/epsilon2)
    xcurr = alpha2*sin(pi*xcurr)
    xcurr2 = alpha2*sin(pi*xcurr2)

plt.plot(xs)
plt.ylabel('(xcurr2-xcurr)/epsilon2')
plt.xlabel('n')
    
```

```python
N = 12
xcurr = .4
alpha2 = .9
epsilon2 = .0001
xcurr2 = xcurr + epsilon2
xs = []
for i in range(N):
    if i > 5:
        xs.append(log(e,abs(xcurr2-xcurr)/epsilon2)/N)
    else:
        xs.append(0)
    xcurr = alpha2*sin(pi*xcurr)
    xcurr2 = alpha2*sin(pi*xcurr2)

plt.plot(xs)
plt.xlabel('N')
plt.axis([6,12,0.01,.04])
```
