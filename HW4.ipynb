```python
import seaborn as sns; sns.set()
from math import pi

N = 100
x = 0
x_data = []
alpha_data = []
alpha_bin = .001
alpha_tot= 1000
x_tot = 10
x_bin = 1 / x_tot
for alpha in range(400, alpha_tot):
    for x in range(1, x_tot):
        alpha_data.append(float(alpha_bin) * (alpha))
        xcurr = x * x_bin        
        for i in range(N):
            if xcurr < .5:
                xcurr = 2*float(alpha_bin) * (alpha)*xcurr
            else:
                xcurr = 2 * float(alpha_bin) * (alpha) * (1-xcurr)
        
        x_data.append(xcurr)

plt.scatter(alpha_data, x_data, s=pi / 2)
plt.ylabel('x')
plt.xlabel('alpha')
```

Before alpha equals .5, we see values of x will go to 0. From .5 to .6, we see that x approaches two values, and afterwards, the alpha values produce chaotic results for x.
